```python
from __future__ import print_function
import numpy as np
import pandas as pd
np.random.seed(1337)  # for reproducibility

from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Dense, Flatten, LSTM, Conv1D, MaxPooling1D, Dropout, Activation
from keras.layers.embeddings import Embedding
from keras.utils.np_utils import to_categorical
from keras.preprocessing import sequence
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import SimpleRNN, GRU
from sklearn.model_selection import train_test_split
```

```python
batch_size = 500

df = pd.read_csv('preprocessed.csv')
tokenizer = Tokenizer()
tokenizer.fit_on_texts(df['text'].astype('str'))
```

```python
sequences = tokenizer.texts_to_sequences(df['text'].astype('str'))
X = pad_sequences(sequences,maxlen= 380,padding="post")
print(X[0])
print(X[1])
Y = to_categorical(df['label'])
print(Y)
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
model_lstm  = Sequential()
model_lstm.add(Embedding(277701, 190, input_length=380))
model_lstm.add(LSTM(190, dropout=0.2, recurrent_dropout=0.2))
model_lstm.add(Dense(3 , activation='tanh'))
model_lstm.compile(loss='categorical_hinge', optimizer='Nadam', metrics=['categorical_accuracy'])

model_lstm.fit(X,Y,batch_size=batch_size,nb_epoch=3)
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.tokenize import RegexpTokenizer
df=pd.read_csv("test_data.csv")
print(df)
nltk.download('stopwords')
ni = nltk.corpus.stopwords.words('english')

stopwords = set(ni)

stopwords.remove("not")
stopwords.remove("against")
stopwords.remove("no")

def preprocess(x):
    if(type(x)==str):
        x = re.sub('[^a-z\s]', '', x.lower())
        x = re.sub(r'[^\w\s]', "", x)
        x = [w for w in x.split() if w not in set(stopwords)]
        return x     




g=[]
for i in range(0,len(df)):
    y = preprocess(df['text'][i])
    g.append(y)

count = 0
array_of_array=[]
for i in g:
    t = " ".join(i)
    array_of_array.append(t)

print(array_of_array)

newd = pd.DataFrame(array_of_array)
print(newd)
tokenizer = Tokenizer()
tokenizer.fit_on_texts(newd[0].astype('str'))
sequences = tokenizer.texts_to_sequences(newd[0].astype('str'))
X = pad_sequences(sequences,maxlen= 380)
y_pred =model_lstm.predict(x=X) 
```

```python
l=[]
for j in range(len(y_pred)):
    if(list(y_pred[j]).index(max(y_pred[j])) == 0):
        l.append(0)
    elif(list(y_pred[j]).index(max(y_pred[j])) == 1):
        l.append(1)
    elif(list(y_pred[j]).index(max(y_pred[j])) == 2):
        l.append(2)

pd.DataFrame(l).to_csv("result1.csv")
print(l)
```
